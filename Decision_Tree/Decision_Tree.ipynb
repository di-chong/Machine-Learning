## 读取csv文件，文件没有表头，直接就是数据，并且所有特征都在一起

```python
data_csv = pd.read_csv('../dataset/German credit/german.csv', header=None, names=['all_features'])
print(data_csv.shape)
```

## 将所有特征分割开来，最后一个特征是标签

```python
data_split = data_csv['all_features'].str.split(expand=True)
data = data_split.iloc[:, :20] 
label = data_split.iloc[:, -1] 
label = pd.DataFrame(label) 
data, label
```

## 对于缺失值：1.数据多可删除；2.可以均值（众数）等方法填充

### 当决策树算法处理连续值特征时，它会考虑所有可能的划分点，并选择能够最大化信息增益（或基尼不纯度减少、均方误差减少等，取决于你使用的具体目标函数）的那个划分点。
### 这意味着算法会自动找到一个阈值，将特征空间划分为两部分，并基于这个划分来构建树的节点。

## 将输入数据全部转化为数值型
### 如果是数值字符----->数值类型
### 如果是字母这种字符---->按0-n分类---->数值类型

```python
# 对目标变量-质量等级进行编码处理
label['target'].replace({'1':0,'2':1},inplace=True)
label.info()
```

```python
# 数值字符转化为数值型
# 如果您确定列中没有无法解析的值，或愿意忽略它们（例如，通过删除含有NaN的行），  
  
# 例如，删除含有NaN的行：  
# df = df.dropna(subset=['A', 'B'])  
# 然后再进行转换：  
data.loc[:, 'Duration in month'] = data['Duration in month'].astype(int)  # 如果确定都是整数  
data.loc[:, 'Credit amount'] = data['Credit amount'].astype(int)
data.loc[:, 'Rate of income'] = data['Rate of income'].astype(int)
data.loc[:, 'Present residence'] = data['Present residence'].astype(int)
data.loc[:, 'Age'] = data['Age'].astype(int)
data.loc[:, 'Number of credits'] = data['Number of credits'].astype(int)
data.loc[:, 'Number of dependents'] = data['Number of dependents'].astype(int)
```

```python
data.info()
```

```python
# 对类别型特征进行独热编码  
from sklearn.preprocessing import OneHotEncoder
categorical_features = ['Status of account', 'Credit history', 'Purpose', 'Credit amount',
               'Savings account', 'employment', 'Personal status', 'Other debtors',
                'Property', 'Other installment', 'Housing',
               'Job',  'Telephone', 'foreign worker']  
encoder = OneHotEncoder(sparse=False)  
encoded = encoder.fit_transform(data[categorical_features])  
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names(input_features=categorical_features)) 
encoded_df.info()
```

## 划分数据集

```python
from sklearn import model_selection
train_x, test_x, train_y, test_y = model_selection.train_test_split(encoded_df, label, test_size=0.3)
train_x
# train_x = pd.get_dummies(train_x, columns=['category_feature']) 
```
